# Question 2: Difference between List and Tuple in Python

## 📌 Definition
Both **list** and **tuple** are Python data structures used to store multiple values.

## 🔍 Main Difference
- A **list** is **mutable** → it can be changed after creation.
- A **tuple** is **immutable** → it cannot be changed after creation.

## 📋 Comparison Table

| Feature | List | Tuple |
|---|---|---|
| Syntax | `[ ]` | `( )` |
| Mutable? | Yes ✅ | No ⚠️ |
| Performance | Slightly slower | Slightly faster |
| Use case | When data may change | When data should stay fixed |

## 🌍 Real-life example
- **List:** Shopping list, because you may add/remove items.
- **Tuple:** Coordinates of a location `(x, y)`, which usually stay fixed.

## ✅ Summary
Use a **list** when you need flexibility.  
Use a **tuple** when you want fixed, protected data.


```python
# List example
fruits_list = ["apple", "banana", "mango"]
fruits_list.append("orange")
print("List after adding an item:", fruits_list)

# Tuple example
fruits_tuple = ("apple", "banana", "mango")
print("Tuple:", fruits_tuple)
```

```python
# Visualization: comparing mutability concept
categories = ["List length", "Tuple length"]
values = [len(fruits_list), len(fruits_tuple)]

plt.figure(figsize=(6, 4))
plt.bar(categories, values)
plt.title("List vs Tuple: Example Length Comparison")
plt.ylabel("Number of items")
plt.show()
```

# Question 5: Check whether a number is even or odd

## 📌 Principle
A number is:
- **Even** if it is divisible by 2
- **Odd** if it is not divisible by 2

We use the modulus operator `%`.

- If `number % 2 == 0` → even
- Otherwise → odd

## 🌍 Real-life example
- Even number of chairs can be split into equal pairs.
- Odd number leaves one extra.


```python
number = 7

if number % 2 == 0:
    print(number, "is Even ✅")
else:
    print(number, "is Odd ⚠️")
```

```python
# Visualization: even and odd numbers from 1 to 10
numbers = np.arange(1, 11)
labels = ["Even" if n % 2 == 0 else "Odd" for n in numbers]
y = [1 if n % 2 == 0 else 0 for n in numbers]

plt.figure(figsize=(8, 4))
plt.scatter(numbers, y, s=100)
plt.yticks([0, 1], ["Odd", "Even"])
plt.title("Even vs Odd Numbers")
plt.xlabel("Number")
plt.ylabel("Category")
plt.grid(True)
plt.show()
```

# Question 8: Remove duplicate values from a list using a set

## 📌 Definition
A **set** stores only unique values.

## 🧠 Principle
If we convert a list into a set:
- duplicate items are automatically removed

## 🌍 Example
Input list:
`[1, 2, 2, 3, 4, 4, 5]`

Output after using set:
`[1, 2, 3, 4, 5]`

## ⚠️ Important note
A set does not preserve order in the same way a list does.


```python
numbers_with_duplicates = [1, 2, 2, 3, 4, 4, 5]
unique_numbers = list(set(numbers_with_duplicates))

print("Original list:", numbers_with_duplicates)
print("After removing duplicates:", unique_numbers)
```

```python
# Visualization: duplicate frequencies
numbers_with_duplicates = [1, 2, 2, 3, 4, 4, 5]
counts = Counter(numbers_with_duplicates)

plt.figure(figsize=(8, 4))
plt.bar(list(counts.keys()), list(counts.values()))
plt.title("Frequency Before Removing Duplicates")
plt.xlabel("Number")
plt.ylabel("Count")
plt.show()
```

# Question 9: Function that returns the square of a number

## 📌 Concept
The square of a number is:

\[
\text{square}(n) = n \times n
\]

## 🧠 Principle
We create a function that:
1. takes one input number
2. multiplies it by itself
3. returns the result


```python
def square(number):
    return number * number

print("Square of 6 is:", square(6))
```

```python
# Visualization: squares of numbers from 1 to 10
x = np.arange(1, 11)
y = x ** 2

plt.figure(figsize=(8, 4))
plt.plot(x, y, marker='o')
plt.title("Squares of Numbers from 1 to 10")
plt.xlabel("Number")
plt.ylabel("Square")
plt.grid(True)
plt.show()
```

# Question 10: Count how many times a number appears in a list

## 📌 Concept
We can use:
- the list method `.count()`
- or the `Counter` class from `collections`

## 🌍 Example list
`[2, 3, 4, 2, 5, 2]`

The number `2` appears **3 times**.


```python
example_list = [2, 3, 4, 2, 5, 2]
target = 2

count_result = example_list.count(target)

print("List:", example_list)
print(f"The number {target} appears {count_result} times.")
```

```python
# Alternative approach with Counter
example_list = [2, 3, 4, 2, 5, 2]
counts = Counter(example_list)
print("Frequencies:", counts)
```

```python
# Visualization: frequency of each value in the example list
example_list = [2, 3, 4, 2, 5, 2]
counts = Counter(example_list)

plt.figure(figsize=(8, 4))
plt.bar(list(counts.keys()), list(counts.values()))
plt.title("Frequency of Numbers in the Example List")
plt.xlabel("Number")
plt.ylabel("Count")
plt.show()
```
